### Duplicate Handling

No duplicate records were found in the dataset.

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"], dayfirst=True)
```

### Data Type Conversion

The OrderDate column was converted from string to datetime format to enable time-based analysis.

```python
df.info()
```

```python
df["Category"] = df["Category"].str.strip()
df["Brand"] = df["Brand"].str.strip()
df["PaymentMethod"] = df["PaymentMethod"].str.strip()
df["OrderStatus"] = df["OrderStatus"].str.strip()
df["City"] = df["City"].str.strip()
df["State"] = df["State"].str.strip()
df["Country"] = df["Country"].str.strip()
```

### Categorical Data Cleaning

Leading and trailing spaces were removed from categorical columns such as Category, Brand, PaymentMethod, OrderStatus, City, State, and Country to ensure consistency.

```python
Q1 = df["TotalAmount"].quantile(0.25)
Q3 = df["TotalAmount"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["TotalAmount"] < lower) | (df["TotalAmount"] > upper)]

outliers.shape
```

### Outlier Detection and Treatment

Outliers in the TotalAmount column were identified using the IQR method.

Records outside the acceptable range were removed to improve data quality and ensure accurate analysis.

```python
df = df[(df["TotalAmount"] >= lower) & (df["TotalAmount"] <= upper)]
```

### Final Dataset After Cleaning

After removing outliers, the dataset size was reduced from 100,000 rows to 98,640 rows, ensuring better data quality.
