## Dataset Overview

This dataset contains Amazon sales transaction data with 100,000 records and 20 features, representing customer orders, product details, and sales metrics.

It includes:
- Order details (OrderID, OrderDate)
- Customer information (CustomerID, CustomerName)
- Product details (ProductID, ProductName, Category, Brand)
- Sales metrics (Quantity, UnitPrice, Discount, Tax, ShippingCost, TotalAmount)
- Transaction details (PaymentMethod, OrderStatus)
- Location data (City, State, Country)
- Seller information (SellerID)

The dataset is suitable for analyzing sales performance, customer behavior, product trends, and regional insights.

## Business Understanding

The objective of this analysis is to answer key business questions:

- Which products and categories generate the most revenue
- Customer purchasing behavior
- Most popular payment methods
- Sales trends over time
- Regional performance (city/state/country)

This analysis will support data-driven decision-making for improving sales strategies and customer engagement.

## Data Quality Issues

1. The `OrderDate` column was initially in string format and has been converted to datetime for proper time-based analysis.
2. No missing values were found across any columns in the dataset, indicating complete data.
3. No duplicate records were detected, ensuring data uniqueness.
4. Some categorical columns such as `Category`, `Brand`, and `PaymentMethod` may require standardization (e.g., consistent naming formats), which will be handled in Day 2.
5. Numerical columns like `UnitPrice`, `Tax`, `ShippingCost`, and `TotalAmount` show a wide range of values, indicating potential presence of outliers. These will be analyzed and treated in Day 2.

Overall, the dataset is clean with no major structural issues, but requires further preprocessing for consistency and outlier handling.
